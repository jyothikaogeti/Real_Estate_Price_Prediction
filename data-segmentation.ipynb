The Dataframe contains information about Apartments, Independent Houses, Villas, Residental Plots.
Divide the DataFrame into separate DataFrames based on property types (such as Apartments, Independent Houses, Villas, and Residential Plots) and save each one as a CSV file

```python
# Function to extract property_type
def extract_property_type(property_desc):
    # Ckeck if 'BHK' or 'RK' is in the description and split accordingly
    if isinstance(property_desc, str) and 'BHK' in property_desc:
        parts = property_desc.split('BHK', 1)
        property_type = parts[1].strip()
    elif isinstance(property_desc, str) and 'RK' in property_desc:
        parts = property_desc.split('RK', 1)
        property_type = parts[1].strip()
    else:
        property_type = property_desc if isinstance(property_desc, str) else 'Unknown'
        
    return property_type
```

```python
df.insert(1, "property_type", df['name'].apply(extract_property_type))
```

- Flats: Apartment, Independent Floor, Studio Apartment
- Houses: Villa, Independent House
- Land: Residential Plot

```python
# Divide the data frame based on property type
flats = ['Apartment', 'Independent Floor']
houses = ['Independent House', 'Villa']
land = ['Residential Plot']

flats_df = df[df['property_type'].isin(flats)]
houses_df = df[df['property_type'].isin(houses)]
land_df = df[df['property_type'].isin(land)]
```

```python
print(f"Total Flats: {len(flats_df)}")
print(f"Total Independent Houses: {len(houses_df)}")
print(f"Total Residental Plots: {len(land_df)}")
```

### Save the dataframes to csv files.

```python
# Save each divided dataframe to CSV files
flats_df.to_csv('../data/raw_data/hyderabad/flats.csv', index=False)
houses_df.to_csv('../data/raw_data/hyderabad/houses.csv', index=False)
land_df.to_csv('../data/raw_data/hyderabad/land.csv', index=False)
```
